```python
def create_table(database_file):
    import sqlite3

    conn = sqlite3.connect(database_file)
    cursor = conn.cursor()
    cursor.execute(f'''
        
        create table if not exist company_tax_code (
            tax_code text
            , company_name text
            , update_time date
        )        
    ''')

    conn.commit()
    conn.close()

def extract_data(database, table_name= 'company_tax_code'):
    import sqlite3
    import pandas as pd

    conn = sqlite3.connect(database)

    try:
        df = pd.read_sql(f'select * from {table_name} ', con= conn)
        return df
    except Exception as e:
        print(e)

    conn.close()

def truncate_data(table_name):
    import sqlite3

    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    cursor.execute(f'''
        delete from {table_name}
    ''')
    conn.commit()
    conn.close()
```

```python
create_table('sample_database.db - sample')
```

```python
extract_data(database= 'database.db', table_name= 'company_tax_code')
```
